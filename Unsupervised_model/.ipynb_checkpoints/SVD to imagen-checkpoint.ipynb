```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from Unsupervised_model import SVD
```

```python
# Editar la imagen del usuario
img = Image.open("Imagen.jpeg")
img_arr = np.array(img)

# Convert to grayscale
img_gray = np.mean(img_arr, axis=2)
```

```python
# Apply SVD
U, S, V = np.linalg.svd(img_gray)

# Visualize singular values
plt.plot(S)
plt.xlabel("Number of singular values")
plt.ylabel("Singular value")
plt.show()
```

```python
# Reconstruct the image using progressively more singular values
n_values = [1, 5, 10, 20, 50, 100, 200, 500]
fig, axs = plt.subplots(2, 4, figsize=(12, 6))
for i, n in enumerate(n_values):
    img_reconstructed = U[:, :n] @ np.diag(S[:n]) @ V[:n, :]
    axs[i//4][i%4].imshow(img_reconstructed, cmap="gray")
    axs[i//4][i%4].set_title(f"{n} singular values")
    axs[i//4][i%4].axis("off")
plt.tight_layout()
plt.show()
```

```python
n_values = [1, 5, 10, 20, 50, 100, 200, 500]
mse_values = []
for n in n_values:
    img_reconstructed = U[:, :n] @ np.diag(S[:n]) @ V[:n, :]
    mse = np.mean((img_gray - img_reconstructed)**2)
    mse_values.append(mse)

plt.plot(n_values, mse_values)
plt.xlabel("Number of singular values")
plt.ylabel("Mean squared error")
plt.show()
```
